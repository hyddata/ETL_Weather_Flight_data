```python
# Install dependencies (uncomment if needed)
# !pip install sqlalchemy
# !pip install pymysql

import pandas as pd
import requests
from sqlalchemy import create_engine, Integer, text
from sqlalchemy.types import String, Float, DateTime

# Step 1: Define cities and fetch metadata
cities = ['Berlin', 'Hamburg', 'Munich']
cities_json = []

for city in cities:
    url = 'https://api.api-ninjas.com/v1/city'
    header = {"X-Api-Key": "qClItv2fg4hGDiVLf6zTNw==rLE3xLfQa0uwZFD0"}
    querystring = {"name": city}

    city_request = requests.get(url, headers=header, params=querystring)
    city_json = city_request.json()

    if city_json:
        city_info = city_json[0]
        city_info["input_city"] = city
        city_info["country_name"] = city_info["country"]
        cities_json.append(city_info)

cities_df = pd.DataFrame(cities_json)
cities_df = cities_df.drop(['is_capital', 'region', 'input_city', 'country_name'], axis=1)
cities_df = cities_df.rename(columns={'name': 'city_name'})

# Step 2: Add city_id
cities_df.reset_index(inplace=True)
cities_df.rename(columns={'index': 'city_id'}, inplace=True)

# Step 3: Define weather fetch function
def get_city_weather(city_name, lat, lon, api_key):
    url = (
        f"https://api.openweathermap.org/data/2.5/forecast?"
        f"lat={lat}&lon={lon}&appid={api_key}&units=metric"
    )
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Failed to fetch weather for {city_name}: {response.status_code}")
        return pd.DataFrame()

    data = response.json()
    records = []

    for entry in data.get("list", []):
        record = {
            "city_name": city_name,
            "latitude": lat,
            "longitude": lon,
            "population": data["city"]["population"],
            "country": data["city"]["country"],
            "forecast_time": entry["dt_txt"],
            "outlook": entry["weather"][0]["description"],
            "temperature": entry["main"]["temp"],
            "wind_speed": entry["wind"]["speed"],
            "rain": entry.get("rain", {}).get("3h", 0),
            "snow": entry.get("snow", {}).get("3h", 0),
        }
        records.append(record)

    return pd.DataFrame(records)

# Step 4: Fetch weather data for all cities
API_KEY = '610294ad80d078d5d6f7a3dc20dfaef9'
all_weather = []

for _, row in cities_df.iterrows():
    city_name = row["city_name"]
    lat = row["latitude"]
    lon = row["longitude"]
    city_id = row["city_id"]

    weather_df = get_city_weather(city_name, lat, lon, API_KEY)
    if not weather_df.empty:
        weather_df["city_id"] = city_id
        all_weather.append(weather_df)

weather_data = pd.concat(all_weather, ignore_index=True)
weather_data["forecast_time"] = pd.to_datetime(weather_data["forecast_time"])
weather_data.drop_duplicates(subset=["city_id", "forecast_time"], inplace=True)

# Step 5: Connect to MySQL and insert data
schema = "weather_flight_data"
host = "127.0.0.1"
user = "root"
password = "USERPASSWORD"
port = 3306
connection_string = f'mysql+pymysql://{user}:{password}@{host}:{port}/{schema}'
engine = create_engine(connection_string)

# Define SQL types
dtype_mapping = {
    "city_id": Integer(),
    "city_name": String(50),
    "population": Integer(),
    "latitude": Float(),
    "longitude": Float(),
    "country": String(50),
    "forecast_time": DateTime(),
    "outlook": String(100),
    "temperature": Float(),
    "wind_speed": Float(),
    "rain": Float(),
    "snow": Float()
}

# Write to SQL
weather_data.to_sql(
    name="weather_condition",
    con=engine,
    if_exists="replace",  # use "append" to accumulate
    index=False,
    dtype=dtype_mapping
)

# Step 6: Set composite primary key (city_id )
with engine.connect() as conn:
    conn.execute(text("ALTER TABLE weather_condition ADD PRIMARY KEY (city_id, forecast_time);"))
```

```python
# ==================== VISUALIZATION SECTION ====================
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates

# Set style
sns.set(style="whitegrid", font_scale=1.1)
plt.rcParams['figure.figsize'] = (14, 8)

# Ensure forecast_time is datetime (already done, but just in case)
weather_data["forecast_time"] = pd.to_datetime(weather_data["forecast_time"])

# Round time to nearest 3 hours for cleaner plotting (optional)
weather_data['date'] = weather_data['forecast_time'].dt.date
weather_data['time'] = weather_data['forecast_time'].dt.strftime('%H:%M')

# ==============================================================
# 1. Temperature Forecast Over Time (One plot per city + combined)
# ==============================================================
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

# Combined temperature plot
sns.lineplot(
    data=weather_data,
    x="forecast_time",
    y="temperature",
    hue="city_name",
    marker="o",
    linewidth=2.5,
    ax=axes[0]
)
axes[0].set_title("5-Day Temperature Forecast Comparison", fontsize=16, fontweight='bold')
axes[0].set_ylabel("Temperature (C)")
axes[0].set_xlabel("Date & Time")
axes[0].legend(title="City")
axes[0].xaxis.set_major_formatter(DateFormatter("%m-%d %H:%M"))
axes[0].tick_params(axis='x', rotation=45)

# Individual city temperature plots
for i, city in enumerate(cities):
    if i < 3:  # Only 3 cities
        city_data = weather_data[weather_data["city_name"] == city]
        sns.lineplot(
            data=city_data,
            x="forecast_time",
            y="temperature",
            marker="o",
            color=sns.color_palette("husl", 3)[i],
            linewidth=3,
            ax=axes[i+1]
        )
        axes[i+1].set_title(f"{city} - Temperature Forecast", fontsize=14, fontweight='bold')
        axes[i+1].set_ylabel("Temperature (C)")
        axes[i+1].set_xlabel("Date & Time")
        axes[i+1].xaxis.set_major_formatter(DateFormatter("%a %H:%M"))
        axes[i+1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ==============================================================
# 2. Max/Min/Avg Temperature Comparison (Bar Chart)
# ==============================================================
temp_summary = weather_data.groupby("city_name")["temperature"].agg(
    ["mean", "max", "min"]
).round(1).reset_index()

temp_summary_melted = temp_summary.melt(id_vars="city_name", var_name="Metric", value_name="Temperature (C)")

plt.figure(figsize=(10, 6))
sns.barplot(
    data=temp_summary_melted,
    x="city_name",
    y="Temperature (C)",
    hue="Metric",
    palette="coolwarm"
)
plt.title("Temperature Statistics by City (Next 5 Days)", fontsize=16, fontweight='bold')
plt.legend(title="Metric")
plt.tight_layout()
plt.show()

# ==============================================================
# 3. Wind Speed & Rainfall (Dual Axis)
# ==============================================================
fig, ax1 = plt.subplots(figsize=(14, 7))

for city in cities:
    city_data = weather_data[weather_data["city_name"] == city].sort_values("forecast_time")
    ax1.plot(city_data["forecast_time"], city_data["wind_speed"], label=f"{city} Wind", linestyle="--", alpha=0.7)

ax2 = ax1.twinx()
for city in cities:
    city_data = weather_data[weather_data["city_name"] == city].sort_values("forecast_time")
    ax2.bar(city_data["forecast_time"], city_data["rain"], alpha=0.4, width=0.1, label=f"{city} Rain" if city == cities[0] else "")

ax1.set_xlabel("Time")
ax1.set_ylabel("Wind Speed (m/s)", color='tab:blue')
ax2.set_ylabel("Rain in last 3h (mm)", color='tab:green')
ax1.set_title("Wind Speed & Rainfall Forecast (Next 5 Days)", fontsize=16, fontweight='bold')

# Combine legends
lines, labels = ax1.get_legend_handles_labels()
bars, bar_labels = ax2.get_legend_handles_labels()
ax1.legend(lines + bars, labels + bar_labels, loc="upper left")

ax1.xaxis.set_major_formatter(DateFormatter("%m-%d %H:%M"))
plt.setp(ax1.xaxis.get_majorticklabels(), rotation=45)
plt.tight_layout()
plt.show()

# ==============================================================
# 4. Heatmap: Temperature by Day and Time of Day
# ==============================================================
weather_data['hour'] = weather_data['forecast_time'].dt.hour
weather_data['day'] = weather_data['forecast_time'].dt.date

pivot_temp = weather_data.pivot_table(
    values='temperature',
    index='city_name',
    columns=['day', 'hour'],
    aggfunc='mean'
)

plt.figure(figsize=(16, 6))
sns.heatmap(pivot_temp, cmap="YlOrRd", annot=True, fmt=".1f", linewidths=.5, cbar_kws={'label': 'Temperature (C)'})
plt.title("Temperature Heatmap (Hourly Average per Day)", fontsize=16, fontweight='bold')
plt.xlabel("Date           Hour of Day (UTC)")
plt.ylabel("City")
plt.tight_layout()
plt.show()

# ==============================================================
# Optional: Save plots automatically
# ==============================================================
# os.makedirs("weather_plots", exist_ok=True)
# plt.savefig("weather_plots/temperature_comparison.png", dpi=200, bbox_inches='tight')

print("All visualizations generated successfully!")
```

```python
def get_airports(latitudes, longitudes):
    headers = {
        "X-RdAPI-Key": "ADD API",
        "X-RapidAPI-Host": "aerodatabox.p.rapidapi.com"
    }
    querystring = {"withFlightInfoOnly": "true"}
    all_airports = []

    for lat, lon in zip(latitudes, longitudes):
        url = f"https://aerodatabox.p.rapidapi.com/airports/search/location/{lat}/{lon}/km/50/16"
        response = requests.get(url, headers=headers, params=querystring)
        if response.status_code == 200:
            data = response.json()
            airports = pd.json_normalize(data.get('items', []))
            if not airports.empty:
                airports['latitude'] = lat
                airports['longitude'] = lon
                all_airports.append(airports)
        else:
            print(f"Failed to fetch airports for ({lat},{lon}): {response.status_code}")

    return pd.concat(all_airports, ignore_index=True) if all_airports else pd.DataFrame()
```

```python
latitudes = cities_df['latitude'].tolist()
longitudes = cities_df['longitude'].tolist()

airports_df = get_airports(latitudes, longitudes)

# Optional: keep only relevant columns

airports_df = airports_df.merge(cities_df[['city_id','city_name']], left_on='municipalityName', right_on='city_name', how='left')
airports_df=airports_df[['icao','iata','city_id']]
airports_df
```

```python
# Define SQL types
dtype_airports = {
    "icao": String(10),
    "iata": String(10),
    "city_id": Integer(),
}


# Write to SQL
airports_df.to_sql(
    name="airports",
    con=engine,
    if_exists="replace",  # or "append"
    index=False,
    dtype=dtype_airports
)

with engine.connect() as conn:
    conn.execute(text("ALTER TABLE airports ADD PRIMARY KEY (icao);"))

```

```python
import requests
from datetime import datetime, timedelta
import pytz
import pandas as pd

def tomorrows_flight_arrivals(airports_df):
    """
    Fetch tomorrow's flight arrivals for a list of ICAO codes.
    airports_df: DataFrame with column 'ICAO'
    Returns: DataFrame with standardized schema
    """

    # API key from environment
    api_key = "ADD API"

    berlin_tz = pytz.timezone("Europe/Berlin")
    today = datetime.now(berlin_tz).date()
    tomorrow = today + timedelta(days=1)

    # Split into two 12-hour windows 
    times = [["00:00", "11:59"], ["12:00", "23:59"]]

    headers = {
        "X-RapidAPI-Key": api_key,
        "X-RapidAPI-Host": "aerodatabox.p.rapidapi.com"
    }

    querystring = {
        "withLeg": "true",
        "direction": "Arrival",
        "withCancelled": "false",
        "withCodeshared": "true",
        "withCargo": "true",
        "withPrivate": "true",
        "withLocation": "false"
    }

    list_for_arrivals_df = []

    for icao in airports_df["icao"]:
        print(f"Fetching flights for: {icao}")

        for start, end in times:
            url = (
                f"https://aerodatabox.p.rapidapi.com/flights/airports/icao/"
                f"{icao}/{tomorrow}T{start}/{tomorrow}T{end}"
            )

            response = requests.get(url, headers=headers, params=querystring)

            if response.status_code == 200:
                flights_resp = response.json()

                if "arrivals" in flights_resp:
                    arrivals_df = pd.json_normalize(flights_resp["arrivals"])
                    arrivals_df["arrival_airport_icao"] = icao
                    


                   

                    list_for_arrivals_df.append(arrivals_df)
                else:
                    print(f"No 'arrivals' key for {icao}")
            else:
                print(f"Error fetching {icao}: {response.status_code}")

    if list_for_arrivals_df:
        return pd.concat(list_for_arrivals_df, ignore_index=True)
    else:
        return pd.DataFrame()  # empty if no results
```
